```python
import os
mypath = os.getcwd()
```

```python
for folder, sub_folders, files in os.walk(mypath):
    
    print ('Currently looking at folder: ')
    print(folder)
    
    print('Subfolder here are: ')
    print(sub_folders)
    
    print('\n')
    
    for f in files:
        print(f)
    print('\n')
```
